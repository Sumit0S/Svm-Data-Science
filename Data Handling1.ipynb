```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df["Ticket"].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # capturing numeric data

df['cabin_cat']= df['Cabin'].str[0] # capturing catagoric data
```

```python
df['Ticket_num1'] = pd.to_numeric(df['Ticket'],errors='coerce',downcast='integer') # capturing numeric data

df['Ticket_cat2']= np.where(df['Ticket'].isnull(),df['Ticket'],np.nan) # capturing numeric data
```

```python
df.head()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])
```
